Data Wrong Type

```python
df['TransactionNo'] = df['TransactionNo'].astype('object')
df.info()
```

```python
df_clean = df.copy()
df_clean
```

## **Transformation**

```python
df_group = df_clean.groupby(['TransactionNo', 'Date']).agg({'ProductName': lambda x: ', '.join(x)}).reset_index()
df_group
```

```python
df_group['ProductList'] = df_group['ProductName'].str.split(', ')
df_group = df_group.reset_index(drop=True)
```

```python
df_group
```

```python
mlb = MultiLabelBinarizer()
basket_sets = pd.DataFrame(mlb.fit_transform(df_group['ProductList']), columns=mlb.classes_, index=df_group['TransactionNo'])
basket_sets = basket_sets.reset_index(drop=True)
```

```python
basket_sets
```

```python
item = basket_sets.sum()
items = item.sort_values(ascending=True)
items
```

## **Modelling**

```python
frequent_itemsets = apriori(basket_sets, min_support=0.02, use_colnames=True)

print("Frequent 1-itemset: ")
frequent_1_itemsets = frequent_itemsets[frequent_itemsets['itemsets'].apply(len) == 1].copy()
frequent_1_itemsets.loc[:, 'frekuensi'] = frequent_1_itemsets['support'] * 19790
frequent_1_itemsets['frekuensi'] = frequent_1_itemsets['frekuensi'].round().astype(int)
frequent_1_itemsets = frequent_1_itemsets[['itemsets','frekuensi','support']]
frequent_1_itemsets_sorted = frequent_1_itemsets.sort_values(by='support', ascending=False)

frequent_1_itemsets_sorted
```

```python
pd.set_option('display.max_colwidth', None)  # Menampilkan semua lebar kolom
pd.set_option('display.max_columns', None)   # Menampilkan semua kolom

print("Frequent 2-itemset: ")
frequent_2_itemsets = frequent_itemsets[frequent_itemsets['itemsets'].apply(len) == 2].copy()
frequent_2_itemsets.loc[:, 'frekuensi'] = frequent_2_itemsets['support'] * 19790
frequent_2_itemsets['frekuensi'] = frequent_2_itemsets['frekuensi'].round().astype(int)
frequent_2_itemsets = frequent_2_itemsets[['itemsets','frekuensi','support']]
frequent_2_itemsets_sorted = frequent_2_itemsets.sort_values(by='support', ascending=False)

frequent_2_itemsets_sorted
```

## **Pembentukan Aturan Asosiasi**

```python
pd.set_option('display.max_colwidth', None)  # Menampilkan semua lebar kolom
pd.set_option('display.max_columns', None)   # Menampilkan semua kolom

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
rules['frekuensi antecedents'] = rules['antecedent support'] * 19790
rules['frekuensi consequent'] = rules['consequent support'] * 19790
rules['frekuensi antecedents+consequent'] = rules['support'] * 19790
rules['frekuensi antecedents'] = rules['frekuensi antecedents'].round().astype(int)
rules['frekuensi consequent'] = rules['frekuensi consequent'].round().astype(int)
rules['frekuensi antecedents+consequent'] = rules['frekuensi antecedents+consequent'].round().astype(int)
rules = rules[['antecedents','consequents', 'frekuensi antecedents', 'frekuensi consequent', 'frekuensi antecedents+consequent', 'antecedent support','consequent support', 'support','confidence','lift']]
rules = rules.sort_values(['confidence', 'lift'], ascending=[False, False])
rules
```
